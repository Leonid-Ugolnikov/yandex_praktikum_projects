## Предобработка данных

### Обработка пропусков

```python
display(df.duplicated().sum())#посчитаем количество дубликатов
```

Дубликатов нет, двигаемся дальше.

```python
display(df.isna().sum())#считаем количество пропусков в таблице
```

```python
df[['living_area',
    'kitchen_area',
    'ceiling_height',
    'total_area',
    'rooms',
    'locality_name',
    'balcony']].corr()
#Применим корреляцию Пирсона для нахождения наибольшей зависимости
```

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('locality_name')['ceiling_height'].transform('median'))
#меняем пропущенные значения на медиану расчитанную по названию населенного пункта
df['floors_total'] = df['floors_total'].fillna(df['floor'])
df['balcony'] = df['balcony'].fillna(0)#заменим пропущенные значения на 0
#df['kitchen_area'] = df['kitchen_area'].fillna(0)#заменим пропущенные значения на 0
df.living_area = df.living_area.fillna(df.total_area - df.kitchen_area)#меняем пропущенные значения на разницу площади
df.living_area = df.living_area.fillna(df.groupby('total_area')['living_area'].transform('median'))
#меняем пропущенные значения на медиану расчитанную по названию населенного пункта
df.kitchen_area = df.kitchen_area.fillna(df.total_area - df.living_area)#меняем пропущенные значения на разницу площади
df.kitchen_area = df.kitchen_area.fillna(df.groupby('total_area')['kitchen_area'].transform('median'))
#меняем пропущенные значения на медиану расчитанную по названию населенного пункта


display(df.isna().sum())#проверим количество пропусков после замены пропущенных значений
```

#### Вывод

Пропущенные значения встречаются более чем в половине столбцов таблицы, больше всего их в колонках 'is apartment'(20924), 'park_nearest'(15620), 'ponds_nearest'(14589), 'balcony'(11519).
В столбцах 'floors_total' и 'locality_name' количество пропущенных значений менее 5%, значит эти столбцы можно исключить из обработки пропущенных значений. Пропущенные значения в столбце 'ceiling_height' мы заполним медианой, значение привяжем к 'locality_name', для столбца 'balcony' и 'kitchen_area' пропущенне значения заменим на 0, поскольку вероятнее всего пропущенные значения означают их отсутствие, например кухни могут быть не указанны в квартирах студиях, либо в коммунальных квартирах. В столбцах 'is_apartment', 'airports_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest' и 'living_area' мы не будем менять пропущенные значения, поскольку данных для этого слишком мало. Для столбца 'days_exposition' нецелесообразно заменять пропущенные значения поскльку, возможно квартира еще в продаже, по этой причине не снята с публикации. 

### Изменение типов данных

```python
df.info()#смотрим информацию о типах данных колонок
df.head(10)#выводим первые 10 строк для изучения данных таблицы
```

```python
df= df.round({ 'days_exposition':0,
               'ponds_nearest':0,
               'ponds_around3000':0,
               'parks_nearest':0,
               'parks_around3000':0,
               'last_price':0,
               'floors_total':0,
               'cityCenters_nearest':0,
               'balcony':0,
               'airports_nearest':0
                })
#округляем значения столбцов, для дальнейшей обработки в целочисленное значение
df = df.astype({'ponds_nearest':'Int64',
                'ponds_around3000':'Int64',
                'parks_nearest':'Int64',
                'parks_around3000':'Int64',
                'last_price':'Int64',
                'floors_total':'Int64',
                'days_exposition':'Int64',
                'cityCenters_nearest':'Int64',
                'balcony':'Int64',
                'airports_nearest':'Int64'
                })
#заменим вещественные значения на целочисленные
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
#изменим тип данных в столбце с датой на тип datetime
df.info()#смотрим информацию о типах данных колонок
df.head(10)#выводим первые 10 строк для изучения данных таблицы
```

#### Вывод

Выше мы привели значения некоторых столбцов к целочисленным значениям, а именно, в столбцах  'floors_total', 'balcony', 'ponds_around3000', 'parks_around3000', 'days_exposition' мы изменили тип данных на целочисленный, поскольку данные значения не могут быть дробными. В столбцах 'cityCenters_nearest', 'airports_nearest', 'ponds_nearest', 'parks_nearest' указанны расстояния в метрах, по этой причине мы отбросили дробную часть, предварительно округлив значения. Так же в столбце с датой и временем 'first_day_exposition' изменим тип данных на datetime.

## Расчёты и добавление результатов в таблицу

```python
df['price_square_meter'] = (df['last_price']//df['total_area']).astype('int64')
#добавляем столбец со стоимостью кв метра
df['date_week'] = pd.DatetimeIndex(df['first_day_exposition']).weekday# добавляем столбец с днем
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month# добавляем столбец с месяцем
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year# добавляем столбец с годом

def floor(df):
    if df['floor'] == 1:
        return 'первый'
    elif df['floor'] == df['floors_total']:
        return 'последний'
    else:
        return 'другой'
#функция для описания этажа
df['apartment_floor'] = df.apply(floor,axis=1)#создание столбца и применение функции

df['kitchen_share'] = (df['kitchen_area']/df['total_area']).round(2)
#добавляем столбец отношение площади кухни к общей
df['residental_share'] = (df['living_area']/df['total_area']).round(2)
#добавляем столбец отношение жилой площади к общей площади 
df.head()
```

### Вывод

Мы добавили столбцы со стоимостью квадратного метра, с днем, годом и месяцем продажи, написали функцию, которая добавляет столбец с описанием этажа (первый, последний и другой), а также добавили столбцы с зависимостью кухонной и жилой площади от общей, предварительно округлив значения до сотых.

### Исследование площади, количества комнат, высоты потолков, стоимости и сроков продажи квартир

```python


def end_of_the_chart(column):
    describe_column = df[column].describe()
    iqr = describe_column[6]-describe_column[4]
    left_end_chart = describe_column[4]-1.5*iqr
    right_end_chart = describe_column[6]+1.5*iqr
    if right_end_chart > describe_column[7]: right_end_chart = describe_column[7]
    if left_end_chart < describe_column[3]: left_end_chart = describe_column[3]
    return [left_end_chart.round(2), right_end_chart.round(2)]
# написали функцию для определения "краев" диаграммы boxplot
display("Числовое описание данных и построение boxplot и гистограммы по стобцу 'total_area'")
print('Границы диаграммы boxplot:',end_of_the_chart('total_area'))
df.boxplot(column='total_area', figsize=(12,6))#строим диаграмму размаха boxplot
plt.ylim(-10,150)
plt.show()
display(df['total_area'].describe())# получаем числовое описание данных
df['total_area'].hist(bins=100,range=(0, 170),figsize=(12,6))#строим гистограмму
plt.show()

display("Числовое описание данных и построение boxplot и гистограммы по стобцу 'last_price'")
print('Границы диаграммы boxplot:', end_of_the_chart('last_price'))
df.boxplot(column='last_price',figsize=(12,6))#строим диаграмму размаха boxplot
plt.ylim(10000, 15000000)
plt.show()
display(df['last_price'].describe())# получаем числовое описание данных
df['last_price'].hist(bins=150, range=(10000, 15000000),figsize=(12,6))#строим гистограмму
plt.show()

display("Числовое описание данных и построение boxplot и гистограммы по стобцу 'rooms'")
print('Границы диаграммы boxplot:',end_of_the_chart('rooms'))
df.boxplot(column='rooms',figsize=(12,6))#строим диаграмму размаха boxplot
plt.ylim(0, 10)
plt.show()
display(df['rooms'].describe())# получаем числовое описание данных
df['rooms'].hist(bins=150, range=(0,10),figsize=(12,6))#строим гистограмму
plt.show()

display("Числовое описание данных и построение boxplot и гистограммы по стобцу 'ceiling_height'")
print('Границы диаграммы boxplot:',end_of_the_chart('ceiling_height'))
df.boxplot(column='ceiling_height',figsize=(12,6))#строим диаграмму размаха boxplot
plt.ylim(2, 3)
plt.show()
display(df['ceiling_height'].describe())# получаем числовое описание данных
df['ceiling_height'].hist(bins=150, range=(2, 3),figsize=(12,6))#строим гистограмму
plt.show()

display("Числовое описание данных и построение boxplot и гистограммы по стобцу 'days_exposition'")
print('Границы диаграммы boxplot:',end_of_the_chart('days_exposition'))
df.boxplot(column='days_exposition',figsize=(12,6))#строим диаграмму размаха boxplot
plt.ylim(0, 550)
plt.show()
display(df['days_exposition'].describe())# получаем числовое описание данных
df['days_exposition'].hist(bins=150, range=(0, 550),figsize=(12,6))#строим гистограмму
plt.show()
```

### Удаление аномальных значений

```python
print(f"Площадь жилья: {end_of_the_chart('total_area')}",
      f"Цена: {end_of_the_chart('last_price')}",
      f"Количество комнат: {end_of_the_chart('rooms')}",
      f"Высота потолков: {end_of_the_chart('ceiling_height')}",
      f"Дни продажи: {end_of_the_chart('days_exposition')}", sep='\n')
#Для удобства, используем ранее написанную функцию, по которой увидим диапазон нормальных значений
#выведем значения на экран с помощью print
```
